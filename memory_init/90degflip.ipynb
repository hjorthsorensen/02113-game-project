```python
def rotate_90_clockwise(matrix):
    """Rotates a 2D matrix 90 degrees clockwise."""
    return [list(row) for row in zip(*matrix[::-1])]

def rotate_mem_file(input_file, output_file):
    # 1. Read all lines from the .mem file
    with open(input_file, 'r') as f:
        # Strip whitespace/newlines and ignore comments or empty lines
        lines = [line.strip() for line in f if line.strip() and not line.startswith('//')]
    
    if len(lines) != 1024:
        raise ValueError(f"Expected exactly 1024 lines for a 32x32 image (1 pixel per line), but found {len(lines)}.")

    # 2. Reconstruct the 32x32 pixel matrix
    # Since it's 1 pixel per line, every 32 lines forms one row of the matrix
    matrix = []
    for i in range(0, 1024, 32):
        row = lines[i:i+32]
        matrix.append(row)
    
    # 3. Rotate the matrix 90 degrees clockwise
    rotated_matrix = rotate_90_clockwise(matrix)
    
    # 4. Flatten the 32x32 matrix back into 1024 individual lines
    rotated_lines = []
    for row in rotated_matrix:
        for pixel in row:
            rotated_lines.append(pixel)
            
    # 5. Write the 1024 individual 7-bit lines back to the file
    with open(output_file, 'w') as f:
        for line in rotated_lines:
            f.write(line + '\n')
            
    print(f"Success! Generated exactly {len(rotated_lines)} lines in '{output_file}'.")

# --- Execution ---
rotate_mem_file('backtile_init_9.mem', 'output_rotated.mem')
```
